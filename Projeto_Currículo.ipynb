## Funções de Extração de Texto

### Subtarefa:
Criar funções Python para extrair texto de arquivos .docx e .pdf, já que o ChatGPT precisa do conteúdo textual para análise.


Raciocínio:
Preciso definir duas funções em Python para extrair texto de arquivos .docx e .pdf, conforme especificado nas instruções. Essas funções irão importar as bibliotecas necessárias (Document de docx e PdfReader de PyPDF2) e implementar a lógica para extração de texto de cada tipo de arquivo. Vou incluir ambas as funções em um único bloco de código para concluir a subtarefa de criação das funções de extração de texto.

```python
from docx import Document
from PyPDF2 import PdfReader

def extract_text_from_docx(filepath):
    """
    Extracts text from a .docx file.
    """
    doc = Document(filepath)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

def extract_text_from_pdf(filepath):
    """
    Extracts text from a .pdf file.
    """
    full_text = []
    with open(filepath, 'rb') as file:
        reader = PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            full_text.append(page.extract_text())
    return '\n'.join(full_text)

print("Functions 'extract_text_from_docx' and 'extract_text_from_pdf' defined.")
```

## Chamar API do ChatGPT para Análise e Melhorias

### Subtarefa:
Enviar o texto extraído do currículo para a API do ChatGPT para obter uma versão melhorada e sugestões de estrutura e aparência.


Raciocínio:
A subtarefa exige chamar a API do ChatGPT para análise e melhorias. Isso envolve importar a biblioteca openai, definir a chave da API, criar um cliente, construir um prompt detalhado com o texto do currículo extraído, enviar a solicitação para a API e, em seguida, processar a resposta para separar o conteúdo melhorado das sugestões estruturais.

Todos esses passos serão consolidados em um único bloco de código para cumprir os requisitos da subtarefa.

```python
import openai
import os

from google.colab import userdata

# --- INSTRUÇÕES IMPORTANTES ---
# 1. Obtenha sua chave de API do OpenAI em: https://platform.openai.com/account/api-keys
# 2. Recomenda-se carregar a chave de uma variável de ambiente por segurança.
#    Se você está rodando no Colab e não quer configurar variáveis de ambiente,
#    pode substituí-la diretamente, mas tenha cuidado para não compartilhar seu notebook
#    com a chave exposta.
#    Exemplo: openai_api_key = "SUA_CHAVE_AQUI"

# Carrega a chave de API dos segredos do Colab
openai_api_key = userdata.get('OPENAI_API_KEY')

# Se a chave ainda não estiver definida (por exemplo, se o secret não existir ou estiver vazio),
# você pode adicionar uma verificação ou instrução aqui, mas o userdata.get já deve resolver.

if not openai_api_key:
    print("ERRO: Por favor, forneça sua chave de API do OpenAI nos segredos do Colab (nome: OPENAI_API_KEY).")
    # Inicializa com strings vazias para evitar erros subsequentes se a chave estiver faltando
    improved_resume_content = ""
    structure_suggestions = ""
else:
    client = openai.OpenAI(api_key=openai_api_key)

    # Carregar o prompt do arquivo externo
    try:
        with open('resume_prompt_template.txt', 'r', encoding='utf-8') as f:
            prompt_template = f.read()
        # Inserir o texto do currículo no template do prompt
        prompt_content = prompt_template.format(resume_text=resume_text)
    except FileNotFoundError:
        print("ERRO: O arquivo 'resume_prompt_template.txt' não foi encontrado. Certifique-se de que a célula anterior foi executada com sucesso.")
        prompt_content = ""
        improved_resume_content = ""
        structure_suggestions = ""
    except Exception as e:
        print(f"ERRO ao ler o arquivo do prompt: {e}")
        prompt_content = ""
        improved_resume_content = ""
        structure_suggestions = ""

    if prompt_content:
        print("Enviando solicitação ao ChatGPT...")

        try:
            chat_completion = client.chat.completions.create(
                model="gpt-4o", # Ou "gpt-3.5-turbo" se preferir um modelo mais rápido e mais barato
                messages=[
                    {"role": "system", "content": "Você é um especialista em recrutamento e escrita de currículos. Você fornecerá um currículo aprimorado e sugestões de estrutura e aparência, separando-os claramente com os marcadores especificados."},
                    {"role": "user", "content": prompt_content}
                ]
            )

            full_response = chat_completion.choices[0].message.content
            print("Resposta do ChatGPT recebida. Processando...")

            # Dividir a resposta usando os marcadores
            improved_resume_content = ""
            structure_suggestions = ""

            if "### Conteúdo Aprimorado:" in full_response and "### Sugestões de Estrutura e Aparência:" in full_response:
                parts = full_response.split("### Sugestões de Estrutura e Aparência:")
                if len(parts) > 1:
                    improved_resume_section = parts[0]
                    structure_suggestions = parts[1].strip()

                    if "### Conteúdo Aprimorado:" in improved_resume_section:
                        improved_resume_content = improved_resume_section.split("### Conteúdo Aprimorado:")[1].strip()
                    else:
                        # Fallback if the first marker is not at the very beginning of the section
                        improved_resume_content = improved_resume_section.strip()
                else:
                    print("Aviso: O marcador para 'Sugestões de Estrutura e Aparência' foi encontrado, mas a divisão falhou. Conteúdo completo atribuído ao currículo aprimorado.")
                    improved_resume_content = full_response
                    structure_suggestions = "Não foi possível extrair sugestões de estrutura e aparência automaticamente devido à formatação inesperada."
            else:
                print("Aviso: Os marcadores esperados ('### Conteúdo Aprimorado:' e '### Sugestões de Estrutura e Aparência:') não foram encontrados na resposta do ChatGPT. Toda a resposta será considerada como conteúdo aprimorado.")
                improved_resume_content = full_response
                structure_suggestions = "Não foi possível extrair sugestões de estrutura e aparência automaticamente. Por favor, verifique a resposta bruta do ChatGPT."


            print("\n--- Conteúdo Aprimorado (Primeiros 500 caracteres) ---")
            print(improved_resume_content[:500])
            print("\n--- Sugestões de Estrutura e Aparência (Primeiros 500 caracteres) ---")
            print(structure_suggestions[:500])

        except openai.APIError as e:
            print(f"Erro na API do OpenAI: {e}")
            print("Certifique-se de que sua chave de API está correta e que você tem créditos suficientes.")
            improved_resume_content = ""
            structure_suggestions = ""
        except Exception as e:
            print(f"Ocorreu um erro inesperado ao chamar a API ou processar a resposta: {e}")
            improved_resume_content = ""
            structure_suggestions = ""
```

## Gerar Novo Arquivo DOCX com Conteúdo Melhorado

### Subtarefa:
Com base no conteúdo textual melhorado retornado pelo ChatGPT, criar um novo arquivo .docx. Este arquivo incluirá o texto aprimorado com uma formatação básica. As sugestões de aparência do ChatGPT serão apresentadas separadamente, pois a aplicação automática de layouts complexos não será possível neste passo.

Raciocínio:
Preciso criar um novo arquivo .docx usando a biblioteca python-docx. O improved_resume_content da resposta do ChatGPT será escrito nesse novo documento.

Também irei salvar as structure_suggestions separadamente, pois elas devem ser aplicadas manualmente pelo usuário. Vou incluir uma formatação básica de parágrafos para manter o texto legível e salvar o arquivo com um nome distinto, indicando que se trata da versão melhorada.

Por fim, exibirei as sugestões de estrutura para o usuário.

```python
from docx import Document
from docx.shared import Inches

def create_docx_with_improved_content(content, filename="curriculo_aprimorado.docx"):
    """
    Cria um novo arquivo DOCX com o conteúdo fornecido.
    """
    document = Document()

    # Adicionar o conteúdo linha por linha
    # Isso pode ser refinado para formatar cabeçalhos, bullet points, etc., se o ChatGPT formatar bem.
    for line in content.split('\n'):
        if line.strip().startswith('**') and line.strip().endswith('**'): # Simple check for bold titles
            document.add_heading(line.strip('*').strip(), level=2)
        elif line.strip().startswith('- '):
            document.add_paragraph(line.strip(), style='List Bullet')
        else:
            document.add_paragraph(line.strip())

    document.save(filename)
    return filename

if improved_resume_content:
    output_docx_filename = create_docx_with_improved_content(improved_resume_content)
    print(f"Novo arquivo DOCX '{output_docx_filename}' criado com o conteúdo aprimorado.")

    print("\n--- Sugestões de Estrutura e Aparência (Completas) ---")
    print(structure_suggestions)
    print("-----------------------------------------------------")
else:
    print("Não há conteúdo aprimorado para gerar o arquivo DOCX.")
```
